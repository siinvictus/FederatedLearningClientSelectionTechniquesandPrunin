```python
from collections import defaultdict
import os

import numpy as np
import re
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def read_files(data_dir):
    hps = []
    files = os.listdir(data_dir)
    files = [f for f in files if f.endswith('.csv')]
    for i,f in enumerate(files):
        file_path = os.path.join(data_dir, f)
        tmp= f.split(':')
        hp = {}
        for i,param in enumerate(f.split('_')):
            vals= param.split(':')
            if i !=0:
                if i!= len(f.split('_'))-1:
                    hp[vals[0]]=eval(vals[1])
                else:
                    hp[vals[0]]=eval(vals[1].split('.csv')[0])
                
        
        data=pd.read_csv(file_path)
        hp['Train_Loss_AVG'] = data.iloc[:,1].values
        hp['Test_Accuracy'] = data.iloc[:,2].values

        hps.append(hp)

    return hps 

```

```python
files = read_files('Results')
stats = pd.DataFrame(files)
stats['Test_Accuracy']= stats['Test_Accuracy'].apply(lambda x: x[0])
del stats['Train_Loss_AVG']
stats.groupby(['angle', stats.index]).sum()
```
